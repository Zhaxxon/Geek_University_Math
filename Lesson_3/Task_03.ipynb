```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = []
y1 = []
y2 = []
r=10
for i in range(-100,101):
    x_= r*i/100
    x.append(x_)
    y1.append(np.sqrt(r**2-x_**2))
    y2.append(-np.sqrt(r**2-x_**2))   

fig, ax = plt.subplots(1)
ax.plot(x,y1,'-b',x,y2,'-b')
plt.axis([-r, r, -r, r])   
ax.set_aspect(1)
plt.title('Окружность')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = []
y1 = []
y2 = []
a=4
b=10
for i in range(-100,101):
    x_= a*i/100
    x.append(x_)
    y1.append(np.sqrt(1-x_**2/a**2)*b)
    y2.append(-np.sqrt(1-x_**2/a**2)*b)
plt.axis([-a, a, -b, b])    
plt.plot(x,y1,'-b',x,y2,'-b')
plt.title('Эллипс')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = []
y1 = []
y2 = []
p=10
for i in range(1000):
    x_= i/1000
    x.append(x_)
    y1.append(np.sqrt(2*p*x_))
    y2.append(-np.sqrt(2*p*x_))   
plt.plot(x,y1,'-b',x,y2,'-b')
plt.title('Гипербола')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
