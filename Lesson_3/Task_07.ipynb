1) Напишите код, который будет переводить полярные координаты в декартовы.

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import math 

def polar2decart(r,alpha):
    return r * np.cos(alpha), r * np.sin(alpha)

polar2decart(1, np.pi/2)
```

2) Напишите код, который будет рисовать график окружности в полярных координатах.

```python
R = 5
x = np.linspace(-R,R,100)
y1 = np.sqrt(R**2 - x**2)
y2 = -np.sqrt(R**2 - x**2)

plt.polar(x, y1, color="green",)
plt.polar(x, y2, color="green",)


plt.show()
```
