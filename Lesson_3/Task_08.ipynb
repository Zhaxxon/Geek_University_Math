1) Решите систему уравнений: y = x^2 – 1; exp(x) + x∙(1 – y) = 1

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import math 
from scipy.optimize import fsolve

def equations(p):
    x,y = p
    return (y - x**2 + 1, np.exp(x) + x*(1-y) - 1)

x1,y1 = fsolve(equations, (3,3))
x1,y1
```

Решите систему уравнений и неравенств: y = x^2 – 1; exp(x) + x∙(1 – y) > 1

```python
x = np.linspace(-10,10,100)
y1 = x**2 - 1

y2 = 1 - (1 - np.exp(x))/ x

plt.plot(x, y1)
plt.plot(x, y2, color="red",)


plt.axis([-5, 5, -2, 20])

plt.grid()
plt.show()
```
