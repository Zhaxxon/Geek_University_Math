Нарисуйте график функции:
y(x) = k∙cos(x – a) + b
для некоторых (2-3 различных) значений параметров k, a, b

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import math 

x = np.linspace(-10,10,100)
k = 1
a = 2
b = 3
y1 = k*np.cos(x - a) + b

k = 2
a = -1
b = 4
y2 = k*np.cos(x - a) + b

plt.plot(x, y1)
plt.plot(x, y2, color="green",)
plt.title('y(x) = k∙cos(x – a) + b')

#plt.axis('equal')

plt.grid()
plt.show()
```
