1) Нарисуйте трехмерный график двух параллельных плоскостей.

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = Axes3D(fig)
X = np.arange(-5,5,0.5)
Y = np.arange(-5,5,0.5)
X,Y = np.meshgrid(X,Y)
Z = 2*X + 5*Y
Z1 = 2*X + 5*Y +15
ax.plot_surface(X,Y,Z)
ax.plot_wireframe(X,Y,Z1, color = 'red')
show()
```

2) Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5,5,0.5)
Y = np.arange(-5,5,0.5)
X,Y = np.meshgrid(X,Y)
Z = np.sqrt(2*X**2 + 3*Y**2)*5
Z1 = -2*X*Y -1
ax.plot_wireframe(X,Y,Z)
ax.plot_wireframe(X,Y,Z1, color = 'red')
show()
```
